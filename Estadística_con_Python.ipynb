```python
import numpy as np
import pandas as pd
import math
```

```python
poblacion = np.random.randint(0, 100, 100000)

muestra_aleatoria = np.random.choice(poblacion, 10000)
```

```python
muestra_aleatoria = np.random.randint(1706, 1789, (5,8))
MIN = np.min(muestra_aleatoria)
MAX = np.max(muestra_aleatoria)
print(f"Valor Minimo: {MIN} y maximo {MAX}")
print('--------------------------------------')
print(muestra_aleatoria)
```

```python
# Sort the data in ascending order
sorted_data = np.sort(muestra_aleatoria)

# Calculate the differences between consecutive values
differences = np.diff(sorted_data)

# Find the smallest non-zero difference
unit = np.min(differences[differences > 0])

print("Unidad de Variación de los datos:", unit)
```

```python
# Número de Intervalos de clase
int_clase = np.round(np.sqrt(muestra_aleatoria.size))
print(f"Número de Intervalos de clase: {int_clase}")
# Amplitud del Intervalo
amp_clase = np.around(((MAX-MIN)/int_clase), decimals=1)
amp_clase = math.modf(amp_clase)[1] + unit
print(f"Amplitud del Intervalo: {amp_clase}")
```

```python
tabla_frecuencias = pd.DataFrame(columns=["L.I.", "L.S.", "Marca de Clase", "L.I. (Exacto)", "L.S. (Exacto)", "Frecuencia Absoluta"])

limites = np.arange(MIN, MAX + amp_clase, amp_clase) #sumamos la amplitud, solo para tener mas valores de los esperados
lim_inf = limites[:int(int_clase)]
lim_sup = limites[1:int(int_clase)+1]

tabla_frecuencias["L.I."] = lim_inf
tabla_frecuencias["L.S."] = lim_sup
tabla_frecuencias["Marca de Clase"] = (tabla_frecuencias["L.I."] + tabla_frecuencias["L.S."]) / 2
tabla_frecuencias["L.I. (Exacto)"] = (tabla_frecuencias["L.I."]-unit/2)
tabla_frecuencias["L.S. (Exacto)"] = (tabla_frecuencias["L.S."]+unit/2)
tabla_frecuencias["Frecuencia Absoluta"] = np.array([np.sum((muestra_aleatoria >= lim_inf[n]) & (muestra_aleatoria < lim_sup[n])) for n in range(lim_inf.size)]) #[LimInf - LimSup)

tabla_frecuencias
```

```python
np.sum(tabla_frecuencias["Frecuencia Absoluta"]) == muestra_aleatoria.size
```

```python
tabla_frecuencias["Frecuencia Absoluta Relativa"] = tabla_frecuencias["Frecuencia Absoluta"] / muestra_aleatoria.size
np.sum(tabla_frecuencias["Frecuencia Absoluta Relativa"])
```

```python
tabla_frecuencias["Frecuencia Acumulada"] = np.array([np.sum(tabla_frecuencias["Frecuencia Absoluta"].iloc[:n].values) for n in range(1,int(int_clase)+1)])
tabla_frecuencias
```

```python
tabla_frecuencias["Frecuencia Acumulada Relativa"] = tabla_frecuencias["Frecuencia Acumulada"] / muestra_aleatoria.size
tabla_frecuencias
```
